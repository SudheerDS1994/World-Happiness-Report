```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style("darkgrid")   # theme has a dark background
sns.set_style({"background-color": "lightgrey"})
# The rcParams dictionary is a global dictionary that contains all of the default settings for Matplotlib.
plt.rcParams["font.size"] = 15   #sets the default font size in Matplotlib to 15 points.
plt.rcParams["figure.figsize"] = (10,7)  #sets the default size of a Matplotlib figure to 6 inches wide and 4 inches high.
plt.rcParams["figure.facecolor"] = "#FFE584"  #FFE584, which is a hexadecimal color code for a light yellow color.
```

```python
def correlation(df,threshold):
    col_corr = set()
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df,0.9)
len(set(corr_features))
```

```python
data_columns = ["Country name","Regional indicator","Ladder score","Logged GDP per capita","Social support","Healthy life expectancy","Freedom to make life choices","Generosity","Perceptions of corruption"]
```

```python
df_new = df[data_columns].copy()
```

```python
df_happy = df_new.rename(columns={
    'Country name': 'country_name',
    'Regional indicator': 'regional_indicator',
    'Ladder score': 'ladder_score',
    'Logged GDP per capita': 'logged_GDP_per_capita',
    'Social support': 'social_support',
    'Healthy life expectancy': 'healthy_life_expectancy',
    'Freedom to make life choices': 'freedom_to_make_life_choices',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'perceptions_of_corruption'
})
```

```python
df_happy.head()
```

```python
df_happy.isnull().sum()
```

```python
# plot between happiness and GDP
plt.rcParams['figure.figsize'] = (15,7)
plt.title('Plot between happiness Score and GDP')
sns.scatterplot(x = df_happy.ladder_score,y = df_happy.logged_GDP_per_capita,hue = df_happy.regional_indicator,s = 200)
plt.legend(loc = 'upper left',fontsize = '10')
plt.xlabel("Happiness Score")
plt.ylabel("GDP per capita")
plt.show()
```

```python
gdp_region = df_happy.groupby("regional_indicator")["logged_GDP_per_capita"].sum()
gdp_region
```

```python
gdp_region.plot.pie(autopct = "%1.1f%%")
plt.title('GDP by region')
plt.ylabel('')
```

```python
# total countries
total_country = df_happy.groupby("regional_indicator")[['country_name']].count()
total_country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
cor = df_happy.corr(method="pearson", numeric_only=True)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the heatmap
sns.heatmap(cor, mask=np.zeros_like(cor), cmap="Blues", square=True, ax=ax)

# Add a title
ax.set_title("Correlation Heatmap")

# Show the plot
plt.show()
```

```python
# corruption in regions

corruptions = df_happy.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruptions
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title("perception of corruption in various regions")
plt.xlabel("Regions",fontsize = 15)
plt.ylabel('Corruption Index',fontsize = 15)
plt.xticks(rotation =30,ha ='right')
plt.bar(corruptions.index,corruptions.perceptions_of_corruption)
```

```python
top_10 = df_happy.head()
bottom_10 = df_happy.tail()
```
